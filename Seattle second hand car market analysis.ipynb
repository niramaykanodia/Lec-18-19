## Identifying Most Valuable Vehicles To Potentially Purchase
**A car Willl be considered valuable if:** 

- Total Miles Done < 40,000
- AGE < 4 Years in or after 2018
- Sub Models : Corolla , cross trek , Ram 1500 , F-150 , Tucson
- If the car is below Market Price

```python
cars_df["price_miles_ratio"] = cars_df["selling_price"]/cars_df["miles"]
cars_df.head()


def get_cars_to_buy(sub_model, cars_df):
    
    # Making a dataframe with all vehicles of the inputted submodel
    sub_model_df = cars_df.loc[(cars_df["miles"]<40000) 
            & (cars_df["year"]>=2018)
            & (cars_df["sub_model"] == sub_model)]
    
    # Calculating the median_price_miles_ratio
    median_price_miles_ratio = sub_model_df.price_miles_ratio.median()
    
    # Calculating the market difference
    sub_model_df["market_difference"] = median_price_miles_ratio - sub_model_df["price_miles_ratio"]

    # Identifying which cars should be purchased
    buy_sub_model_df = sub_model_df.loc[sub_model_df["price_miles_ratio"]<=median_price_miles_ratio]
    
    # challenge make a column that allowes to see how much percentage price miles ratio below market value
    buy_sub_model_df["percentage_below_market"] = buy_sub_model_df["market_difference"]/(median_price_miles_ratio)*100
    
    return buy_sub_model_df.sort_values(["price_miles_ratio"], ascending = True)
```

```python
get_cars_to_buy("Civic", cars_df)
```

```python
# Get All Cars to Buy is a Function that accepts a dataframe of cars
def get_all_cars_to_buy(cars_df):

    # Make Sub_Model_List
    # Which is a list of all unique sub models in cars_df
    sub_model_list = cars_df.sub_model.unique()
    
    # Make Recommendations_df
    # This dataframe will hold all the recommended buys 
    # Initially this will be empty
    recommendations_df = pd.DataFrame()

    # For Loop
    ############################################################
    # For Each Sub Model in the Sub_Model_List
    for sub_model in sub_model_list: 
        
        # Make Best_cars_df has all the best cars to buy for the current sub_model
        best_cars_df = get_cars_to_buy(sub_model, cars_df)
        
        # Add all the best cars to buy to the Recommendations_Df
        recommendations_df = pd.concat([best_cars_df, recommendations_df], ignore_index=True)
        
    # Return Recommendations DF
    return recommendations_df
```

```python
most_valuable_cars_df = get_all_cars_to_buy(cars_df)
most_valuable_cars_df.sort_values(["percentage_below_market"], ascending = False)
```

```python
cars_significantly_below_market_df = most_valuable_cars_df.loc[most_valuable_cars_df["percentage_below_market"]>=15]
cars_significantly_below_market_df.sort_values(["percentage_below_market"], ascending = False)
```
